Plot a histogram of `ConvertedComp.`

Create a scatter plot of `Age` and `WorkWeekHrs.`

## Visualizing composition of data


### Pie Charts


Create a pie chart of the top 5 databases that respondents wish to learn next year. Label the pie chart with database names. Display percentages of each database on the pie chart.


```python
QUERY5 = """
SELECT DatabaseWorkedWith, COUNT(*) AS count
FROM DatabaseWorkedWith
GROUP BY DatabaseWorkedWith
ORDER BY count DESC
"""
# Execute the query and read the results into a DataFrame
top_databases1 = pd.read_sql_query(QUERY5, conn)
print(top_databases1)
# the read_sql_query runs the sql query and returns the data as a dataframe
#df1 = pd.read_sql_query(QUERY5,conn)
#df1.columns
```

```python
QUERY1 = """
SELECT DatabaseDesireNextYear, COUNT(*) AS count
FROM DatabaseDesireNextYear
GROUP BY DatabaseDesireNextYear
ORDER BY count DESC
"""

# Execute the query and read the results into a DataFrame
top_databases = pd.read_sql_query(QUERY1, conn)
print(top_databases)
# Plotting the pie chart
plt.figure(figsize=(8, 8))
plt.pie(top_databases['count'], labels=top_databases['DatabaseDesireNextYear'], autopct='%1.1f%%')
plt.title('Top 5 Databases Respondents Wish to Learn Next Year')
plt.show()
```

### Stacked Charts


Create a stacked chart of median `WorkWeekHrs` and `CodeRevHrs` for the age group 30 to 35.


```python
# your code goes here

import pandas as pd
import matplotlib.pyplot as plt

# Define your SQL query to calculate the median WorkWeekHrs and CodeRevHrs for age group 30 to 35
QUERY2 = """
SELECT Age, 
       AVG(WorkWeekHrs) as MedianWorkWeekHrs,
       AVG(CodeRevHrs) as MedianCodeRevHrs
FROM master 
WHERE Age >= 30 AND Age <= 35
GROUP BY Age
"""

# Execute the query and read the results into a DataFrame
median_hours = pd.read_sql_query(QUERY2, conn)

# Plotting the stacked bar chart
plt.figure(figsize=(10, 6))
plt.bar(median_hours['Age'], median_hours['MedianWorkWeekHrs'], label='Median WorkWeekHrs')
plt.bar(median_hours['Age'], median_hours['MedianCodeRevHrs'], bottom=median_hours['MedianWorkWeekHrs'], label='Median CodeRevHrs')
plt.xlabel('Age')
plt.ylabel('Median Hours')
plt.title('Stacked Chart of Median WorkWeekHrs and CodeRevHrs for Age Group 30 to 35')
plt.legend()
plt.grid(True)
plt.show()
```

### Bar Chart


Create a horizontal bar chart using column `MainBranch.`


```python
# your code goes here

MB = """SELECT MainBranch FROM master"""
plot8 = pd.read_sql_query(MB,conn)
print(plot8)
import pandas as pd
import matplotlib.pyplot as plt

# Define your SQL query to count occurrences of each MainBranch category
QUERY3 = """
SELECT MainBranch, COUNT(*) AS count
FROM master
GROUP BY MainBranch
"""

# Execute the query and read the results into a DataFrame
main_branch_counts = pd.read_sql_query(QUERY3, conn)

# Plotting the horizontal bar chart
plt.figure(figsize=(10, 6))
plt.barh(main_branch_counts['MainBranch'], main_branch_counts['count'])
plt.xlabel('Count')
plt.ylabel('MainBranch')
plt.title('Horizontal Bar Chart of MainBranch')
plt.grid(True)
plt.show()

#sns.barplot(x='MainBranch', data=plot8)
#plt.show()
```
